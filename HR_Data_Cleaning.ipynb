```python
import os
import pandas as pd

os.chdir(r"D:\ASHISH\DATA ANALYST\PROJECTS\HR Project")

df = pd.read_excel("HR Project dataset.xlsx")

df["AttritionNum"] = df["Attrition"].map({"Yes": 1, "No": 0})

df.to_excel("hr_cleaned.xlsx", index=False)
```

```python
df["AgeGroup"] = pd.cut(df["Age"], [0, 25, 35, 50, 100])
df["IncomeBand"] = pd.cut(df["MonthlyIncome"], bins=4)

df["Tenure"] = df["YearsAtCompany"]

# flag employees with no promotion for 3+ years
df["NoPromotion3Plus"] = (df["YearsSinceLastPromotion"] >= 3).astype(int)

df.to_excel("hr_cleaned.xlsx", index=False)

print("Cleaned file saved successfully.")
print(df[["Attrition", "AttritionNum", "AgeGroup", "IncomeBand", "Tenure", "NoPromotion3Plus"]].head())
```
